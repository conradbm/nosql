```python
from cassandra.cluster import Cluster

###
### Create a cluster on our localhost, it expects a list of hosts because of 
###    cassandra's many server architecture purposes
###

cluster = Cluster(['localhost'])
```

```python
###
### Create a session on the cluster
###

session = cluster.connect()
```

```python

###
### Now, create a Keyspace our top leve data structure in Cassandra
###
### CQL is the cassandra query language, its a lot like SQL

session.execute("CREATE KEYSPACE stocks \
                    WITH replication = {'class': 'SimpleStrategy', \
                                        'replication_factor':'1'}")

```

```python
###
### Now we want to connect to our Keyspace (database,schema, ect..)
###
session.set_keyspace('stocks')
```

```python
###
### Create a couple tables for stocks
###

session.execute("""DROP TABLE company""")
session.execute("""DROP TABLE indicator_by_company""")

session.execute("""
                    CREATE TABLE company(
                        company_id text,
                        name_latest text,
                        name_previous text,
                        PRIMARY KEY (company_id)
                    )
                """)

session.execute("""
                    CREATE TABLE indicator_by_company(
                        company_id text,
                        indicator_id text,
                        year2010 bigint,
                        year2011 bigint,
                        year2012 bigint,
                        year2013 bigint,
                        year2014 bigint,
                        year2015 bigint,
                        year2016 bigint,
                        PRIMARY KEY (company_id, indicator_id)
                    )
                """)
```
