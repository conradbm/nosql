```python
import pymongo
```

```python
client = pymongo.MongoClient('localhost', 27017)
db = client['UCI-Database']
collection = db['iris']
```

```python
column_names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "label"]

collection.delete_many({})

with open("iris-shuffled.txt") as in_f:
    for line in in_f:
        row_raw = [l.strip('\n').strip(" ") for l in line.split(",")]
        f1 = map(float, row_raw[:-1])
        row_raw[:-1] = f1
        row_dict = dict(zip(column_names, f1))    
        #print row_raw
        collection.insert_one(row_dict)
        
```

```python
collection.count()
```

```python
for doc in collection.find({'Sepal Length':{"$gt":0}}):
    print(doc['Sepal Width'])
```
